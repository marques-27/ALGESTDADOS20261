```python
#1 classificador de três números

num1 = int(input('digite_o_primeiro_numero'))
num2 = int(input('digite_o_segundo_numero'))
num3 = int(input('digite_o_terceiro_numero'))

if num1 >= num2 and num1 >= num3:
    maior = num1
    if num2 >= num3:
        meio, menor = num2, num3
    else:
        meio, menor = num3, num2
elif num2 >= num1 and num2 >= num3:
    maior = num2
    if num1 >= num3:
        meio, menor = num1, num3
    else:
        meio, menor = num3, num1
else:
    maior = num3
    if num1 >= num2:
        meio, menor = num1, num2
    else:
        meio, menor = num2, num1

print(f'Maior: {maior}, Meio: {meio}, Menor: {menor}')
```

```python
#2 calculadora de gorjeta

valor_conta = float(input('valor_da_conta:'))
nota = int(input('nota_do_serviço(1-5):'))
if valor_conta < 0 or (nota < 1 or nota > 5):
    print('dados_invalidos')
else:
    if nota == 5:
       porcentagem = 0.20
    elif nota == 4:
       porcentagem = 0.15
    elif nota == 3:
       porcentagem =0.10
    else:
       porcentagem = 0.0

    gorjeta = valor_conta * porcentagem
    total = valor_conta + gorjeta
    print(f'Gorjeta: R$ {gorjeta:.2f}')
    print(f'Total: R$ {total:.2f}')


```

```python
#9 conversor de temperatura completo
temp = float(input("Temperatura: "))
unidade = input("Unidade (C/F/K): ").upper()

match unidade:
    case "C":
        temp_celsius = temp
        temp_fahrenheit = (temp * 9/5) + 32
        temp_kelvin = temp + 273.15
    case "F":
        temp_celsius = (temp - 32) * 5/9
        temp_fahrenheit = temp
        temp_kelvin = (temp - 32) * 5/9 + 273.15
    case "K":
        if temp < 0:
            print("Erro: Kelvin não pode ser negativo.")
        else:
            temp_celsius = temp - 273.15
            temp_fahrenheit = (temp - 273.15) * 9/5 + 32
            temp_kelvin = temp
    case _:
        print("Unidade inválida.")

if (unidade in ["C", "F"]) or (unidade == "K" and temp >= 0):
    print(f"Celsius: {temp_celsius:.2f} | Fahrenheit: {temp_fahrenheit:.2f} | Kelvin: {temp_kelvin:.2f}")
```

```python
# 19 Detector de Intervalo com Tolerância

alvo = float(input("Digite o valor alvo: "))
tolerancia = float(input("Digite a tolerância: "))
medido = float(input("Digite o valor medido: "))

if alvo <= 0 or tolerancia <= 0 or medido <= 0:
    print("Erro: Todos os valores devem ser números positivos!")
else:
    limite_inferior = alvo - tolerancia
    limite_superior = alvo + tolerancia
    desvio = medido - alvo

    if medido < limite_inferior:
        classificacao = "Abaixo do limite"
    elif medido > limite_superior:
        classificacao = "Acima do limite"
    else:
        classificacao = "Dentro da faixa"

    print("-" * 30)
    print(f"RESULTADO DA ANÁLISE:")
    print(f"Faixa permitida: {limite_inferior:.2f} até {limite_superior:.2f}")
    print(f"Classificação: {classificacao}")
    print(f"Desvio real: {desvio:+.2f}")
    print("-" * 30)
```

```python
#20 Simulador de Pedágio Inteligente

veiculo = input("Tipo de veículo (moto, carro, caminhao, onibus): ")
tag = input("Possui tag de pagamento automático? (s/n): ")
hora = int(input("Hora atual (0-23): "))

if 0 <= hora <= 23:
    match veiculo:
        case "moto":
            tarifa = 5.00
        case "carro":
            tarifa = 10.00
        case "caminhao":
            tarifa = 20.00
        case "onibus":
            tarifa = 15.00
        case _:
            tarifa = 0

    if tarifa > 0:
        if tag == "s":
            tarifa = tarifa * 0.90

        if (7 <= hora <= 9) or (17 <= hora <= 19):
            tarifa = tarifa * 1.20
            horario_msg = "Horário de Pico (+20%)"
        else:
            horario_msg = "Horário Normal"

        print("-" * 30)
        print(f"TARIFA DETALHADA:")
        print(f"Veículo: {veiculo.capitalize()}")
        print(f"Status: {horario_msg}")
        print(f"Tag aplicada: {'Sim' if tag == 's' else 'Não'}")
        print(f"Valor Final: R$ {tarifa:.2f}")
        print("-" * 30)
    else:
        print("Erro: Veículo inválido.")
else:
    print("Erro: Hora inválida (digite entre 0 e 23).")
```

```python
#21 Calculadora de Energia Elétrica

consumo = float(input("Digite o consumo em kWh: "))

if consumo < 0:
    print("Erro: O consumo não pode ser negativo.")
else:
    if consumo <= 100:
        preco_unitario = 0.50
    elif consumo <= 200:
        preco_unitario = 0.75
    elif consumo <= 300:
        preco_unitario = 1.00
    else:
        preco_unitario = 1.50

    valor_base = consumo * preco_unitario
    valor_final = valor_base

    if consumo > 300:
        economico = input("Optou pelo modo econômico? (s/n): ")
        if economico == "s":
            valor_final = valor_base * 0.90

    print("-" * 30)
    print(f"RESUMO DA CONTA:")
    print(f"Consumo total: {consumo} kWh")
    print(f"Valor sem desconto: R$ {valor_base:.2f}")

    if valor_final < valor_base:
        print(f"Desconto aplicado (10%): R$ {(valor_base - valor_final):.2f}")
        print(f"Valor final com desconto: R$ {valor_final:.2f}")
    else:
        print("Nenhum desconto aplicado.")
    print("-" * 30)
```

```python
#22 Verificador de Intervalo Múltiplo

A = int(input("Digite o intervalo mínimo (A): "))
B = int(input("Digite o intervalo máximo (B): "))
N = int(input("Digite o número (N): "))
D = int(input("Digite o divisor (D): "))

if A >= B or D <= 0:
    print("Erro: A deve ser menor que B e D deve ser positivo e não nulo.")
else:
    if A <= N <= B:
        if N % D == 0:
            print("dentro do intervalo e divisível")
        else:
            print("dentro e não divisível")
    else:
        if N % D == 0:
            print("fora do intervalo e divisível")
        else:
            print("fora e não divisível")
```

```python
#23 Sistema de Pontuação de Jogo

p1 = int(input("Pontuação Jogador 1: "))
p2 = int(input("Pontuação Jogador 2: "))
p3 = int(input("Pontuação Jogador 3: "))

if p1 < 0 or p1 > 1000 or p2 < 0 or p2 > 1000 or p3 < 0 or p3 > 1000:
    print("Erro: As pontuações devem estar entre 0 e 1000.")
else:
    if p1 == p2 == p3:
        primeiro = segundo = terceiro = "Empate Triplo"
        maior_val = menor_val = p1
    elif p1 >= p2 and p1 >= p3:
        primeiro = "Jogador 1"
        maior_val = p1
        if p2 >= p3:
            segundo, terceiro = "Jogador 2", "Jogador 3"
            menor_val = p3
        else:
            segundo, terceiro = "Jogador 3", "Jogador 2"
            menor_val = p2
    elif p2 >= p1 and p2 >= p3:
        primeiro = "Jogador 2"
        maior_val = p2
        if p1 >= p3:
            segundo, terceiro = "Jogador 1", "Jogador 3"
            menor_val = p3
        else:
            segundo, terceiro = "Jogador 3", "Jogador 1"
            menor_val = p1
    else:
        primeiro = "Jogador 3"
        maior_val = p3
        if p1 >= p2:
            segundo, terceiro = "Jogador 1", "Jogador 2"
            menor_val = p2
        else:
            segundo, terceiro = "Jogador 2", "Jogador 1"
            menor_val = p1

    diferenca = maior_val - menor_val

    print("-" * 30)
    print("RANKING FINAL:")
    print(f"1º Lugar: {primeiro} ({maior_val} pts)")
    print(f"2º Lugar: {segundo}")
    print(f"3º Lugar: {terceiro} ({menor_val} pts)")
    print(f"Diferença entre o 1º e o 3º: {diferenca} pts")
    print("-" * 30)
```

```python
#24 Calculadora de Reajuste Salarial

salario_atual = float(input("Digite o salário atual: R$ "))
avaliacao = float(input("Digite o percentual da avaliação (0-100): "))

if salario_atual < 0 or avaliacao < 0 or avaliacao > 100:
    print("Erro: Salário ou avaliação inválidos.")
else:
    if avaliacao >= 90:
        percentual = 15
    elif avaliacao >= 75:
        percentual = 10
    elif avaliacao >= 60:
        percentual = 5
    else:
        percentual = 0

    aumento_reais = salario_atual * (percentual / 100)
    novo_salario = salario_atual + aumento_reais

    aviso_teto = ""
    if novo_salario > 15000:
        novo_salario = 15000
        aviso_teto = "AVISO: Salário ajustado ao teto de R$ 15.000,00."

    print("-" * 35)
    print(f"RESUMO DO REAJUSTE:")
    print(f"Salário anterior: R$ {salario_atual:.2f}")
    print(f"Percentual aplicado: {percentual}%")
    print(f"Valor do aumento: R$ {aumento_reais:.2f}")
    print(f"Novo salário: R$ {novo_salario:.2f}")

    if aviso_teto:
        print(aviso_teto)
    print("-" * 35)
```

```python
# 25
a = float(input("Digite o primeiro ângulo: "))
b = float(input("Digite o segundo ângulo: "))
c = float(input("Digite o terceiro ângulo: "))

if a <= 0 or b <= 0 or c <= 0:
    print("Erro: Todos os ângulos devem ser positivos.")
else:
    soma = a + b + c

    if soma != 180:
        print(f"Triângulo inválido. Soma encontrada: {soma}°")
    else:

        if a < 90 and b < 90 and c < 90:
            tipo_ang = "Acutângulo"
        elif a == 90 or b == 90 or c == 90:
            tipo_ang = "Retângulo"
        else:
            tipo_ang = "Obtusângulo"

        if a == b == c:
            tipo_lado = "Equilátero"
        else:
            if a == b or a == c or b == c:
                tipo_lado = "Isósceles"
            else:
                tipo_lado = "Escaleno"

        print(f"O triângulo é {tipo_ang} e {tipo_lado}.")
```

```python
# 26 calculadora de combustível de viagem

distancia = float(input("Distância em km: "))
consumo = float(input("Consumo do veículo (km/l): "))
preco = float(input("Preço do combustível por litro: "))

if distancia > 0 and consumo > 0 and preco > 0:
    litros_necessarios = distancia / consumo
    custo_total = litros_necessarios * preco

    print(f"Litros necessários: {litros_necessarios:.2f}L")
    print(f"Custo total: R${custo_total:.2f}")


    if custo_total > 500:
        metade_percurso = distancia / 2
    print(f"Alerta: Custo alto! Sugestão de parada no km {metade_percurso:.2f}")
else:
    print("Erro: Todos os valores devem ser positivos.")
```

```python
#27 verificador de senha com nível de segurança

senha = input("Digite a senha: ")
confirmacao = input("Confirme a senha: ")
if senha != confirmacao:
    print("Senhas não conferem")
else:
    comprimento = len(senha)
    if comprimento < 6:
        nivel = "Fraca"
    elif 6 <= comprimento <= 9:
        nivel = "Média"
    else:
        nivel = "Forte"

    if senha.isdigit():
        if nivel == "Forte":
            nivel = "Média"
        elif nivel == "Média":
            nivel = "Fraca"
    print(f"Nível final da senha: {nivel}")
```

```python
#28 simulador de reserva de assento
assento = int(input("Número do assento (1-100): "))
classe = input("Classe (economica, executiva, primeira): ").lower()

if 1 <= assento <= 100:

    match classe:
        case "primeira":

            if 1 <= assento <= 10:
                print(f"Confirmado: Assento {assento} na classe Primeira. Preço: R$1.500,00")
            else:
                print("Assento incompatível com a classe Primeira.")

        case "executiva":
            if 11 <= assento <= 30:
                print(f"Confirmado: Assento {assento} na classe Executiva. Preço: R$890,00")
            else:
                print("Assento incompatível com a classe Executiva.")

        case "economica":
            if 31 <= assento <= 100:
                print(f"Confirmado: Assento {assento} na classe Econômica. Preço: R$350,00")
            else:
                print("Assento incompatível com a classe Econômica.")

        case
            print("Classe inválida.")
else:
    print("Número de assento inválido (deve ser entre 1 e 100).")
```

```python
#29 simulador de batalha por tipos

tipo_atacante = input("Tipo do atacante (fogo, agua, planta, eletrico, terra):")
tipo_defensor = input("Tipo do defensor (fogo, agua, planta, eletrico, terra):")
poder_base = float(input("Poder base do ataque: "))

multiplicador = 1.0


match tipo_atacante:
    case "fogo":

        if tipo_defensor == "planta":
            multiplicador = 2.0
        elif tipo_defensor == "agua":
            multiplicador = 0.5

    case "agua":
        if tipo_defensor == "fogo":
            multiplicador = 2.0
        elif tipo_defensor == "planta":
            multiplicador = 0.5

    case "planta":
        if tipo_defensor == "agua":
            multiplicador = 2.0
        elif tipo_defensor == "fogo":
            multiplicador = 0.5

    case "eletrico":
        if tipo_defensor == "agua":
            multiplicador = 2.0
        elif tipo_defensor == "terra":
            multiplicador = 0.5

    case "terra":
        if tipo_defensor == "eletrico":
            multiplicador = 2.0

    case
    print("Tipo de atacante inválido.")


dano_final = poder_base * multiplicador
print(f"O dano final é: {dano_final:.2f} (Multiplicador: x{multiplicador})")
```

```python
#30  sistema completo de reserva de hotel

tipo_quarto = input("Tipo de acomodação (standard, luxo, suite): ")
diarias = int(input("Quantidade de diárias: "))
fidelidade = input("Possui cartão fidelidade? (s/n): ")


if diarias > 0:

    match tipo_quarto:
        case "standard":
            diaria_base = 150.0
        case "luxo":
            diaria_base = 300.0
        case "suite":
            diaria_base = 500.0
        case
            diaria_base = 0
            print("Tipo de quarto inválido.")

    if diaria_base > 0:

        if 1 <= diarias <= 3:
            desconto_qtd = 0.0
        elif 4 <= diarias <= 7:
            desconto_qtd = 0.10
        else:
            desconto_qtd = 0.20

        desconto_fidelidade = 0.0
        if fidelidade == 's':
            desconto_fidelidade = 0.05


        subtotal = diaria_base * diarias
        valor_desconto_qtd = subtotal * desconto_qtd
        valor_desconto_fid = subtotal * desconto_fidelidade
        total_final = subtotal - valor_desconto_qtd - valor_desconto_fid


        if total_final >= 100:
            print(f"\n--- RESUMO DA RESERVA ---")
            print(f"Quarto: {tipo_quarto.capitalize()}")
            print(f"Valor da Diária: R${diaria_base:.2f}")
            print(f"Total de Diárias: {diarias}")
            print(f"Desconto por tempo: R${valor_desconto_qtd:.2f}")
            print(f"Desconto fidelidade: R${valor_desconto_fid:.2f}")
            print(f"TOTAL FINAL: R${total_final:.2f}")
        else:
            print(f"Erro: O valor total R${total_final:.2f} é inferior ao mínimo de R$100,00.")
else:
    print("Erro: O número de diárias deve ser positivo.")
```
